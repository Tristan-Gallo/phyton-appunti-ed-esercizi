```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

a. creare un dataframe contenente le informazioni riportate qui sotto, relative a quanti prodotti ha
venduto (‘quantity’) e quanto ha venduto in termini di $ (‘total_amount’) ogni membro dello staff
(colonna “last_name”):

```python
lst = {'last_name':['Boyer','Copeland','Daniel','Serrano','Terrel','Vargas'],'quantity':[766,213,812,269,100,114],'total_amount':[643049.37,203549.39,744259.20,253183.68,93851.95,123156.90]}
df = pd.DataFrame(lst)
df
```

b. visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
fig,(ax,ax1) = plt.subplots(1,2)
fig.set_size_inches(15,8)
plt.rcParams.update({"font.size" : 12})
#fig.autofmt_xdate(rotation="vertical")
ax.bar(df["last_name"], df["quantity"])
ax1.bar(df["last_name"], df["total_amount"])
plt.show()
```

c. visualizzi i grafici precedenti con un titolo, gli assi e la griglia.

```python
fig,(ax,ax1) = plt.subplots(1,2)
fig.set_size_inches(15,8)
plt.rcParams.update({"font.size" : 12})
#fig.autofmt_xdate(rotation="vertical")
ax.bar(df["last_name"], df["quantity"])
ax1.bar(df["last_name"], df["total_amount"])
ax.set_xlabel('last_name')
ax.set_ylabel('quantity')
ax1.set_xlabel('last_name')
ax1.set_ylabel('total_amount')
ax1.set_title("A")
ax1.set_title("B")
fig.suptitle('Grafici vendite')
ax.grid()
ax1.grid()
plt.show()
```

d. visualizzi, senza creare nuovi dataframe, il costo medio dei prodotti venduti da ogni dipendente. Il
risultato deve essere questo (prestare attenzione ai valori dell’asse delle ordinate):

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df["last_name"], df["total_amount"]/df["quantity"])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_ylim([800,1100])

fig.suptitle('Istogramma')
plt.show()
```
