```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
lst = {'last_name':['Boyer','Copeland','Daniel','Serrano','Terrel','Vargas'],'orders_number':[185,50,196,58,24,28]}
df = pd.DataFrame(lst)
df
```

b. visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:

```python
fig = plt.figure()
ax = plt.axes()

ax.bar(df['last_name'],df['orders_number'])
```

c. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df['last_name'],df['orders_number'],color='r')
fig.suptitle('nomi e cognomi dei comessi')
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df.sort_values('orders_number',ascending=False)['last_name'],df['orders_number'].sort_values (ascending = False),color = 'r')
```

e. visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

```python
fig = plt.figure()
ax = plt.axes()
ax.barh(df['last_name'],df['orders_number'])
```

```python
df2 = pd.DataFrame()
df2['last_name'] = df['last_name'].sort_values(ascending = True)
df2['orders_perc'] = df['orders_number']/df['orders_number'].sum()*100
df2
```

g. visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
cols = ['c','m','r','g','b','y']
fig = plt.figure(figsize = [7,7])
ax = plt.axes()
ax.pie(df['orders_number'], labels=df['last_name'], colors= cols, startangle=90,autopct = '%1.1f%%')
fig.suptitle('grafico percentuale dipendenti', color= 'k')
plt.show()
```

b. visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
cols = ['c','m','r','g','b','y']
fig = plt.figure(figsize = [11,11])
ax =plt.axes()

ax.pie(df2['orders_perc'], labels=df2['last_name'],  startangle=90, autopct='%1.1f%%')
fig.suptitle('grafico dei dipendenti', color= 'k')

plt.show()
```
